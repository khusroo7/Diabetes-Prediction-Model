```python
import seaborn as sns
from matplotlib import pyplot as plt
import numpy as np
import pandas as pd
```

```python
diabetes_data=pd.read_csv("diabetes.csv")
```

```python
x=diabetes_data.drop(columns='Outcome',axis=1)
```

```python
y=diabetes_data['Outcome']
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=123)
```

```python
print(x_train.shape,y_train.shape)
```

```python
print(x_test.shape,y_test.shape)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
logreg=LogisticRegression()
```

```python
logreg.fit(x_train,y_train)
```

```python
logreg_predict=logreg.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
```

```python
accuracy_score(y_test,logreg_predict)
```

```python
test_data_rs=test_data_np.reshape(1,-1) #-1 lets the system figures out what the additional dimension is.
```

```python
test_predict=logreg.predict(test_data_rs)
```

```python
test_predict #it will give the output based on the values whether the patient is diabetic or not using logistoc regression
```

```python
print(confusion_matrix(y_test,logreg_predict))
```

```python
print(classification_report(y_test,logreg_predict))
```
